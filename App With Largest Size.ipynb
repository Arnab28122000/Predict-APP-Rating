```python
import numpy as np
import pandas as pd 
```

```python
df=pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

## Remove initials like 'M','K','B','G' from Size column

```python
df['Size']=df['Size'].apply(lambda x:str(x).replace('M',''))
df['Size']=df['Size'].apply(lambda x:str(x).replace('K',''))
df['Size']=df['Size'].apply(lambda x:str(x).replace('B',''))
df['Size']=df['Size'].apply(lambda x:str(x).replace('G',''))
```

## Convert size column datatype into numeric,and make NAN which have string value inside

```python
df.Size = pd.to_numeric(df.Size, errors='coerce')
```

```python
df['Size']
```

## Check maximum of size column

```python
df['Size'].max()
```

## All data with size 100.0

```python
df[df['Size']==100.0]
```

## Check total null value in Size column

```python
df['Size'].isnull().sum()
```

## Handling null value with mean

```python
df['Size'].fillna(df['Size'].mean(),inplace=True)
```

```python
df['Size'].isnull().sum()
```

```python
ans=df[['Size','App']][df['Size']==df['Size'].max()]
```
